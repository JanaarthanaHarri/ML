### Q10. Use Seaborn "TIPS" dataset and plot pairplot, jointplot, catplot, displot, and HeatMap

"Comments on the different plots you used and why you used them what observations you made. Regarding the Heatmap, comment on observations you can make with a heat map and do you see any correlation in the data in consideration, which normalisation technique you used."

````python
tips = sns.load_dataset("tips")
tips
````

````python
sns.histplot(data=tips['day'])    #   no of customers are high during saturdays
````

````python
sns.displot(data=tips['total_bill']) 
````

# Pairplot

````python
sns.pairplot(tips, hue = 'day')         # default pairplot plots only the numerical values from the dataset

'''
OBSERVATION:
 
  - There is a positive correlation with tip and total_bill, so we can say the higher the bill the higher the tip.
  - In histograms we can observe saturdays are have more customers and friday has the lowest.
  - Also tip and size has positive relation.
'''
````

````python
sns.pairplot(tips, hue = 'sex') 

'''
OBSERVSTION:
  - No of male customers are high compared to female.
  - High Positive correlation for male to total_bill than female.
'''
````

````python
sns.pairplot(tips, hue = 'smoker')
````

````python
sns.lmplot("total_bill", "tip", hue = "smoker", data = tips);

'''
  Non smoker has high positive correlation than smoker.
'''
````

````python
sns.pairplot(tips, x_vars=["total_bill", "size"], y_vars=["tip"], height=5, aspect=.8, kind="reg");
````

````python
'''
Label encoder is used to change the string values to a corresponding integer values

'''

from sklearn.preprocessing import OneHotEncoder, LabelEncoder
LabelEncoder_X= LabelEncoder()
tips['sex']= LabelEncoder_X.fit_transform(tips['sex'])
tips['smoker']= LabelEncoder_X.fit_transform(tips['smoker'])
tips['time']= LabelEncoder_X.fit_transform(tips['time'])
tips['day']= LabelEncoder_X.fit_transform(tips['day'])
````

### Min Max normalization technique is used

````python
from sklearn.preprocessing import MinMaxScaler
norm = MinMaxScaler().fit(tips)
tips_norm = norm.transform(tips)
````

````python
df = pd.DataFrame(tips_norm)
````

````python
df.columns = ['total_bill','tip','sex','smoker','day','time','size']
````

````python
df         # normalized dataframe
````

````python
sns.pairplot(df)  # NO new important observations with normalized pairplot.
````

# Jointplot

````python
sns.jointplot(x= 'total_bill', y= 'tip', data = tips)

'''
OBSERVATIONS:
  - Has positive relation
  - strength seems to be moderate as some points are scattered.
  - Marginal histograms are right skewed.
  - We can see some outliers in both scatter plot and histograms.
'''
````

````python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg') # Regression line of best fit is drawn.
````

# Catplot

````python
sns.catplot(x='day' , y= 'total_bill' ,data = tips, kind = 'strip') # distribution of total_bill across each day 
````

````python
sns.catplot(x='day' , y= 'total_bill' ,data = tips, kind = 'swarm')

'''
  Similar to stripplot but the points are adjusted so they don't overlap.
'''
````

````python
sns.catplot(x='day' , y= 'total_bill' ,hue = 'time' ,data = tips, kind = 'box')

'''
  Distribution of 'total_bill' for each 'day' and for each 'time' within each 'day'
'''
````

````python
sns.catplot(x='day' , y= 'total_bill' ,hue = 'time' ,data = tips, kind = 'violin')

'''
  Same as boxplot but with kernal density curve of other numerical variable.
'''
````

# Displot

````python
'''
  kdeplot - Plot univariate or bivariate distributions using kernel density estimation.
'''

sns.displot(tips.tip, kde = 'True')
````

````python
sns.displot(data=tips, x="total_bill", y="tip", kind="kde", rug=True) # shows individual observations with marginal rug
````

````python
sns.displot(data=tips, x="tip", hue="day", kind="kde")      # frequency of tip across each day
````

````python
sns.displot(data=tips, x="tip", hue="smoker",col = "sex", kind="kde")   
````

````python
sns.displot(data=tips, x="tip", hue="day", multiple = 'stack')
sns.displot(data=tips, x="total_bill", hue="day", multiple = 'stack')

'''
 Visualization to show subsets on distinct subplots.
'''
````
